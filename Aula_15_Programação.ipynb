```python
mf = np.ceil(10*notas.mean(axis = 1))/10

# Adicionar ao conjunto de dados
notas['MF'] = np.ceil(10*notas.mean(axis = 1))/10
notas
```

```python
# O aluno é considerado aprovado quando sua nota MF >= 6
def auxiliar(nota):
  if nota >= 6:
    saida = 'Aprovado'
  else:
    saida = 'Reprovado'
  return saida

notas['Situação'] = notas['MF'].apply(auxiliar)
notas
```
