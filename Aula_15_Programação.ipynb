```python
dicionario = {'Nomes': nomes, 'Renda_Anual': renda_anual, 'Renda_Mensal': renda_mensal, 'Conclusão': conclusao }

conjunto_de_dados = pd.DataFrame(dicionario)
conjunto_de_dados
```

```python
def auxiliar(conc):
  if conc < 0.25:
    saida = 'baixa'
  elif conc >= 0.25 and conc < 0.50:
    saida = 'media'
  elif conc >= 0.50 and conc < 0.75:
    saida = 'alta'
  else:
    saida = 'excelente'
  return saida

conjunto_de_dados['Faixa_de_Conclusao'] = conjunto_de_dados['Conclusão'].apply(auxiliar)
```

```python
def auxiliar(tam):
    if len(tam) <= 14:
      saida = 'pequeno'
    else:
      saida = 'grande'
    return saida

conjunto_de_dados['Tamanho_do_Nome'] = conjunto_de_dados['Nomes'].apply(auxiliar)
```
